```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob
```

```python
comments=pd.read_csv('vaccination_tweets.csv')
comments
```

```python
comments.head()
```

```python
polarity =[]

for i in comments['text']:
    polarity.append(TextBlob(i).sentiment.polarity)
```

```python
comments['polarity']=polarity
```

```python
comments.head()
```

```python
comments_positive=comments[comments['polarity']==1]
```

```python
comments_positive.head()
```

```python
!pip install wordcloud
```

```python
from wordcloud import WordCloud,STOPWORDS
```

```python
stopwords=set(STOPWORDS)
```

```python
total_comments=' '.join(comments_positive['text'])
```

```python
wordcloud=WordCloud(width=1000,height=500,stopwords=stopwords).generate(total_comments)
```

```python
plt.figure(figsize =(15,5))
plt.imshow(wordcloud)
plt.axis('off')
```

```python
comments_negative =comments[comments['polarity']==-1]
comments_negative
```

```python
total_negcomments=' '.join(comments_negative['text'])
```

```python
wordcloud=WordCloud(width =1000,height=500,stopwords=stopwords).generate(total_negcomments)
```

```python
plt.figure(figsize=(15,5))
plt.imshow(wordcloud)
plt.axis('off')
```
